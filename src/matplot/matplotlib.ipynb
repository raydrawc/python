```python
# 设置格式
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# 绘制曲线
x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
```

```python
# 调整线条颜色 color 简写 c
offsets = np.linspace(0, np.pi, 5)
colors = ['blue', 'g', 'r', 'yellow', 'pink']
for offset, color in zip(offsets, colors) :
    plt.plot(x, np.sin(x-offset), color=color)
```

```python
# 线条风格linestype = ls
# 调整线宽 linewidth = lw
# 标记点样式 marker
# 标记点大小 markersize
```

```python
# xlim ylim 限制最大最小值

x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x))
plt.xlim(-1, 7)
plt.xlim(-0.75, )
```

```python
# axis 设置轴大小
x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x))
plt.axis([-1,8,-2,2]) # xmin xmax ymin ymax
```

```python
x = np.logspace(0,5, 100)
plt.plot(x, np.log(x))
# 输出不同形式
plt.xscale("log")
```

```python
# 散点图
x = np.linspace(0, 2*np.pi, 100)
y = x ** 2
plt.scatter(x, np.sin(x), c = y, cmap="inferno") # cmap颜色映射
```
